```python
from detect_password import *

WORD_LEN = 12
PADDING_CHAR = '^' # Use ^ since it's rarely used in normal word

detector = PasswordDetector('models/rnn_model.h5', word_len=WORD_LEN, padding_char=PADDING_CHAR)
```

```python
normal_words = ['Hello', 'at least 8 character long', 'SELECT * FROM table_name;', '</html>', "col_name", 'field=']
password_words = ['147258', '123456789a', 'blink182', 'michael1', '1qaz2wsx', 'sunshine']
test_words = normal_words + password_words

pd.DataFrame({
    'OriginalWord': list(test_words),
    'IsPassword': [False] * len(normal_words) + [True] * len(password_words),
    'Prediction': detector.detect_passwords(test_words)})
```
